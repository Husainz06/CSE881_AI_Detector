```python
import pandas as pd

raw_data = pd.read_csv("scraped_data.csv")

# randomly selecting 1,000 jobs to be the new dataset
raw_1000 = raw_data.sample(n=1000, random_state=1)
# dropping Unnamed: 0 column
raw_1000 = raw_1000.drop(columns=["Unnamed: 0"])
# reset index
raw_1000 = raw_1000.reset_index(drop=True)
raw_1000.head()
```

```python
# removing '\n' characters from job descriptions because AI generated jobs do not have it
raw_1000['job_summary'] = raw_1000['job_summary'].str.replace('\n', ' ')
# Adding target column with 0 for not AI generated
raw_1000['is_AI'] = 0
raw_1000.head()
```

```python
# saving data
raw_1000.to_csv("raw_1000.csv", index=False)
```

```python
# AI jobs
raw_AI = pd.read_csv("ai_job_postings_1000-2.csv")
raw_AI['is_AI'] = 1
raw_AI.head()
```

```python
# combining datasets
combined = pd.concat([raw_1000, raw_AI], ignore_index=True)
# shuffling raw order to mix up the postings
combined = combined.sample(frac=1, random_state=1).reset_index(drop=True)
combined.head()
# saving combined dataset
combined.to_csv("combined.csv", index=False)
```
