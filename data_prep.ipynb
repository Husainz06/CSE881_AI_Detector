```python
import pandas as pd

raw_data = pd.read_csv("scraped_data.csv")

# randomly selecting 1,500 jobs to be the new dataset
raw_1500 = raw_data.sample(n=1500, random_state=1)
# dropping Unnamed: 0 column
raw_1500 = raw_1500.drop(columns=["Unnamed: 0"])
# reset index
raw_1500 = raw_1500.reset_index(drop=True)
raw_1500.head()
```

```python
# removing '\n' characters from job descriptions because AI generated jobs do not have it
raw_1500['job_summary'] = raw_1500['job_summary'].str.replace('\n', ' ')
# Adding target column with 0 for not AI generated
raw_1500['is_AI'] = 0
raw_1500['source'] = 'human'
raw_1500.head()
```

```python
# saving data
raw_1500.to_csv("raw_1500.csv", index=False)
```

```python
# AI jobs
raw_claude = pd.read_csv("claude_data.csv")
# choose 300 random samples from the dataset
raw_claude = raw_claude.sample(n=300, random_state=1)
raw_claude['is_AI'] = 1
raw_claude['source'] = 'claude ai'
raw_claude = raw_claude.reset_index(drop=True)
raw_claude.head()
```

```python
raw_gemini = pd.read_csv("gemini_data.csv")
raw_gemini['is_AI'] = 1
raw_gemini['source'] = 'gemini ai'
# renaming colums to match the other datasets
raw_gemini = raw_gemini.rename(columns={"jobtitle": "job_title", "joblocation": "job_location", "jobsalary": "job_salary", "jobsummary": "job_summary"})
raw_gemini.head()
```

```python
raw_chatgpt = pd.read_csv("chatgpt_data.csv")
raw_chatgpt['is_AI'] = 1
raw_chatgpt['source'] = 'chatgpt ai'
raw_chatgpt = raw_chatgpt.rename(columns={"jobtitle": "job_title", "joblocation": "job_location", "jobsalary": "job_salary", "jobsummary": "job_summary"})
raw_chatgpt.head()
```

```python
raw_copilot = pd.read_csv("copilot_data.csv")
raw_copilot['is_AI'] = 1
raw_copilot['source'] = 'copilot ai'
raw_copilot.head()
```

```python
raw_preplixity = pd.read_csv("preplixity_data.csv")
raw_preplixity['is_AI'] = 1
raw_preplixity['source'] = 'preplixity ai'
raw_preplixity = raw_preplixity.rename(columns={"jobtitle": "job_title", "joblocation": "job_location", "jobsalary": "job_salary", "jobsummary": "job_summary"})
# dropping extra columns
raw_preplixity = raw_preplixity.drop(columns=["company","jobtype","posted_date"])
raw_preplixity.head()
```

```python
# combining datasets
combined = pd.concat([raw_1500, raw_claude, raw_chatgpt, raw_copilot, raw_gemini, raw_preplixity], ignore_index=True)
# shuffling raw order to mix up the postings
combined = combined.sample(frac=1, random_state=1).reset_index(drop=True)
# saving combined dataset
combined.to_csv("combined.csv", index=False)
```
